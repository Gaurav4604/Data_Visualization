# Understanding factors for Churn in a Telecom Company

# About the Data:

# The dataset is about telecom industry which tells about the number of customers who churned the service. It consists of 3333 observations having 21 variables. We have to predict which customer is going to churn the service.

# Account Length: how long account has been active.
# VMail Message: Number of voice mail messages send by the customer.

# Day Mins: Time spent on day calls.

# Eve Mins: Time spent on evening calls.

# Night Mins: Time spent on night calls.

# Intl Mins: Time spent on international calls.

# Day Calls: Number of day calls by customers.

# Eve Calls: Number of evening calls by customers.

# Intl Calls: Number of international calls.

# Night Calls: Number of night calls by customer.

# Day Charge: Charges of Day Calls.

# Night Charge: Charges of Night Calls.

# Eve Charge: Charges of evening Calls.

# Intl Charge: Charges of international calls.

# VMail Plan: Voice mail plan taken by the customer or not.

# State: State in Area of study.

# Phone: Phone number of the customer.

# Area Code: Area Code of customer.

# Intl Plan: Does customer have international plan or not.

# CustServ Calls: Number of customer service calls by customer.

# Churn : Customers who churned the telecom service or who doesn’t(0=“Churner”, 1=“Non-Churner”)


```python
# 1.Import necessary libraries
import numpy as np
import pandas as pd
```

```python
# 2.Import the data churn data
data = pd.read_csv('Churn.csv')
```

```python
# 3.How do we display the dataframe? Show from both end
data.head(n=5)
```

```python
data.tail(n=5)
```

```python
# 4.Calculate Histogram for time spent on day calls by customers. Is it normally distributed? Justify it!!!
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(data["total day minutes"])

#Yes the data is normally distributed as most data distribution is near the mean of the data
```

```python
# 5.Describe statistically through five statistics. 
data.describe()
```

```python
# 6.How do we categorize the churner and the non-churner for the time spent on day calls(total day minutes)?
import seaborn as sns
sns.boxplot(x='churn', y='total day minutes', data=data)
```

```python
# 7.Find the number of customers who did opt for voice mail plan
data['voice mail plan'].value_counts()[1]
```

```python
# 8.Produce a countplot for the above result
sns.countplot(x='voice mail plan', data=data)
```

```python
# 9 .Create a boxplot for a categorical variable(international plan) and continuous variable(area code).
sns.boxplot(x=data['international plan'], y=data['area code'], data=data)
```

```python
# 10. Create a crosstab for the area code to find the churner or non-churner.
pd.crosstab(data['area code'], data['churn'])
```

```python
# 11. How to pivot information using python for categorical values? Plot one.
pd.pivot_table(data,index=['churn'], aggfunc=len)
```

```python
# 12.Now calculate the total international minutes for all the combinations above. 
pd.pivot_table(data, 'international plan', index=['churn'], aggfunc=len)
```

```python
# 13.How do we understand the correlation between the variables or the columns within the dataframe. Plot one and analyze.
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True,vmin=-1,vmax=1)
```

```python
# 14. Find Standard deviation of total night calls.
print(np.std(data['total night calls']))
```

```python
# 15.Plot a distplot for the above result to look at specifically total night calls.
sns.distplot(data['total night calls'])
```
